```python
#import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
#load football csv data
df = pd.read_csv('/content/drive/MyDrive/Data Analytics/EPL_20_21.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# check is data has null value or not
df.isna().sum()
```

```python
# Add 2 column goal per match and min per match
df['Goal_per_match']= df['Goals']/df['Matches'].astype(float)
df['Goal_per_min'] = df['Goals']/df['Mins'].astype(int)
df.head()
```

```python
# Calculate total goal, total penalty, total attempt
totalGoal = df['Goals'].sum()
totalPenaltyGoals = df['Penalty_Goals'].sum()
totalPenaltyAttempt = df['Penalty_Attempted'].sum()
print(totalGoal, totalPenaltyGoals, totalPenaltyAttempt)
```

```python
# Pie chart show the missed penalty goal, scored penalty goal
plt.figure(figsize=(13,6))
totalPenaltyMiss= totalPenaltyAttempt - totalPenaltyGoals
data = [totalPenaltyMiss,totalPenaltyGoals]
labels = ['Penalty Missed','Penalty Scored']
color = sns.color_palette(('Set2'))
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.title("Percentage of penalty missed and penalty scrored", loc='center', fontsize = 15)
```

```python
#most player from which country

nationality = df.groupby('Nationality').size().sort_values(ascending = False) 
nationality.head(10).plot(kind='bar', figsize = (12,6), color = sns.color_palette("magma"))
plt.title("Top 10 country have the largest number of players", loc = 'center', fontsize = 15)
```

```python
#Top 5 club has the maximum players in the squad
club = df.groupby('Club').size().sort_values(ascending = False) 
club.head(5).plot(kind='bar', figsize = (12,6), color = sns.color_palette("viridis"))
plt.title('Top 5 club has the maximum players in the squad', loc='center');
```

```python
#Top 5 club with the fewest players in their squad
fig, ax = plt.subplots(figsize = (10,5))
club = pd.DataFrame(df.groupby('Club')['Name'].count().sort_values(ascending = True).head(5) )

#club.head(5).plot(kind='bar', figsize = (12,6), color = sns.color_palette("viridis"))
ax.bar(club.index,club['Name'])
plt.xticks(rotation=90) #rotate x value
for index,data in enumerate(list(club['Name'])):
    plt.text(x=index , y =data+1 , s=f"{data}" , fontsize=10)
plt.tight_layout()
plt.title('Top 5 club with the fewest players in their squad',loc="center");
plt.show()
```

```python
#Percentage players by age group ()
# under 20
# between 20 and 25
# between 25 and 30
# below 30
under20 = df[df['Age']<=20]
between20And25 = df[(df['Age'] > 20) & (df['Age'] <= 25)]
between25And30 = df[(df['Age'] > 25 & (df['Age'] <= 30))]
below30 = df[df['Age'] > 30]

data = np.array([under20['Name'].count(), between20And25['Name'].count(), between25And30['Name'].count(), below30['Name'].count()])
labels = ['Under 20', 'Between 20 and 25', 'Between 25 and 30', 'Below 30']
color = sns.color_palette(("Set2"))
plt.pie(data,colors = color, labels = labels, autopct = '%.0f%%')
plt.title("Percentage of players based on age group", fontsize=15, loc = 'center')
```

```python
#Total player has age under 20 per each club
fig, ax = plt.subplots(figsize = (10,5))
under20 = df[df['Age']<20]
totalUnder20 = pd.DataFrame(under20.groupby('Club')['Name'].count().sort_values(ascending = True) )
ax.bar(totalUnder20.index,totalUnder20['Name'])
plt.xticks(rotation=90) #rotate x value
for index,data in enumerate(list(totalUnder20['Name'])):
    plt.text(x=index , y =data+0.2 , s=f"{data}" , fontsize=10)
plt.tight_layout()
plt.title("Total player has age under 20 per each club", fontsize = 15)
plt.show()


```
